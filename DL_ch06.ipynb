### 6.3.2. Logistic Sigmoid and Hyperbolic Tangent

g$ (z) = \sigma(z)$

g$ (z) = \tanh(z)$

$ tanh(z) = 2\sigma(2z) - 1 $


```python
x = np.linspace(-8, 8, 200)
y = np.exp(x) / (np.exp(x) + 1)
plt.plot(x, y)
plt.title("Losigtic Sigmoid")
plt.show()
```

```python
x = np.linspace(-8, 8, 200)
y = np.tanh(x)
plt.plot(x, y)
plt.title("Hyperbolic Tangent")
plt.show()
```

#### 6.4.1 Universal Approximation Properties and Depth

universal approximation theorem 
-  regardless of what function we are trying to learn, we know that a large MLP will be able to represent this function
- However, we are not guaranteed that the training algorithm will be able to learn that function

we can interpret the use of a deep architecture as expressing a belief that the function we want to learn is a computer program consisting of multiple steps, where each step makes use of the previous step’s output

Empirically, greater depth does seem to result in better generalization for a wide variety of tasks
